```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

#Recopilacion de tweets mediante webscraping

query= 'TheBridge_Tech until:2022-05-24 since:2022-01-01'
tweets = []
limit = 5000


for tweet in sntwitter.TwitterSearchScraper(query).get_items():

    # print(vars(tweet))
    # break
    if len(tweets) == limit:
        break
    else:
        tweets.append([tweet.id, tweet.content, tweet.date,tweet.user.id, tweet.username,tweet.likeCount, tweet.retweetCount, tweet.replyCount, tweet.quoteCount])

df = pd.DataFrame(tweets, columns=['Tweet_ID','Tweet_Content', 'Date','User_ID', 'User', 'Likes', 'Retweets', 'Replies', 'quoteCount'])
print(df)
df.to_csv('Data/tweets.csv')
```

# Aplicacion del modelo

```python
import pickle
with open('Data/finished_model.model', "rb") as archivo_entrada:
    model = pickle.load(archivo_entrada);
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer


def signs_tweets(tweet):
    signos = re.compile("(\.)|(\;)|(\:)|(\!)|(\?)|(\¬ø)|(\@)|(\,)|(\")|(\()|(\))|(\[)|(\])|(\d+)(#)")
    return signos.sub(' ', tweet.lower())


def remove_stopwords(df):
    spanish_stopwords = stopwords.words('spanish')
    return " ".join([word for word in df.split() if word not in spanish_stopwords])


def spanish_stemmer(x):
    stemmer = SnowballStemmer('spanish')
    return " ".join([stemmer.stem(word) for word in x.split()])

tweets['Tweet_Content'] = tweets['Tweet_Content'].apply(signs_tweets)
#tweets['Tweet_Content'] = tweets['Tweet_Content'].apply(remove_stopwords)
tweets['Tweet_Content'] = tweets['Tweet_Content'].apply(spanish_stemmer)
tweets
```

```python
# Me guardo los tres comentarios con m√°s relevancia
best3 = tweets.sort_values(by=["Relevancia"], ascending=False).head(3)
best3
```

```python
predictions = model.predict(best3['Tweet_Content'])
best3['Polarity'] = (predictions)
best3
```

Los resultados con Polarity 0 son positivos y los resultados polarity 1 negativos
